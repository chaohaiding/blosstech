```python
def parseECGdata(ecg_data):
    verification_code=ecg_data[0:4]
    result=[]
    #print(verification_code)
    if verification_code=='FDFE':
        # 6character from 4
        for n in range(1, 7):
            hex_data=ecg_data[4+(n-1)*6:4+n*6]
            result_b=int(hex_data[0:2], 16)*65536+int(hex_data[2:4], 16)*256+int(hex_data[4:6], 16)
            result_b/=20971
            result.append(result_b)
        return result
    else:
        return -1
```
